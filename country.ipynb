```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv("country_wise_latest[1].csv")
```

```python
df.loc[df['WHO Region']=='Africa']
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=[50,50])
plt.barh(df['Country/Region'].value_counts().index,df['Country/Region'].value_counts().values,color=('lightblue','pink'),label='state')
plt.title("This bar graph")
plt.xlabel('margin')
plt.ylabel('name')
plt.grid()
plt.legend(loc='upper left')
plt.show()
```

```python
plt.figure(figsize=[50,50])
plt.pie(df['Country/Region'].value_counts().values,autopct='%1.1f%%',labels=df['Country/Region'].value_counts().index )
```
